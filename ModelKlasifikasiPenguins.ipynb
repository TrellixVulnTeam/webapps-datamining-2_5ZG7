```python
for col in encode:
    dummy = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df,dummy], axis=1)
    del df[col]
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
df['species'] = df['species'].apply(target_encode)
```
